```python
dfa=df.copy()
```

```python
dfa = dfa.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb =LabelEncoder()
```

```python
dfa['Nmarried']=lb.fit_transform(dfa['married'])
```

```python
dfa['Nstatus']=lb.fit_transform(dfa['status'])
```

```python
dfa['Ngender']=lb.fit_transform(dfa['gender'])
```

```python
dfa
```

```python
dfas = dfa.drop(columns=['gender','married','status'])
```

```python
dfas
```

```python
x=dfas.drop(columns=['Nstatus'])
y=dfas['Nstatus']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.27,random_state=1133)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
```

```python
y_predict=svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
score=svc.score(x_test,y_test)
score
```
